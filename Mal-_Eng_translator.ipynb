**TASK: Create a simple AI Model capable of translating MALAYALAM to ENGLISH**

I used MarianMT model provided by Hugging Face Transformer;This can directly perform malayalam to english translation

```python
import gradio as gr
from transformers import MarianMTModel, MarianTokenizer
model_name = "Helsinki-NLP/opus-mt-ml-en"  #malayalam to english model name
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)



```

Funtion to Provide a Malayalam sentence that we have to translate and return it's translation

```python
def mal_to_eng_translator(mal_sentence):

    input_tokenizer=tokenizer(mal_sentence,return_tensors='pt',padding=True,truncation=True)
    output_tokenizer=model.generate(**input_tokenizer,num_beams=5,early_stopping=True)
    translation = tokenizer.decode(output_tokenizer[0], skip_special_tokens=True)
    return translation


```

Gradio Interface

```python
iface = gr.Interface(
    fn=mal_to_eng_translator,
    inputs=gr.Textbox(label="Enter Malayalam Sentence"),
    outputs=gr.Textbox(label="English Translation"),
    title="Malayalam to English Translator",
    description="Enter a Malayalam sentence to get its English translation instantly !!"
)

iface.launch(share=True)
```
